### 6. Укажите возраст (Age) самого молодого игрока, у которого хладнокровие (Composure) и реакция (Reactions) превышают 90% от максимального значения, представленного в датасете.

```python
football[(football.Composure > football.Composure.max()*0.9)  & (football.Reactions > football.Reactions.max()*0.9)].Age.min()
```

#### 7. Определите, насколько средняя реакция (Reactions) самых взрослых игроков (т.е. игроков, чей возраст (Age) равен максимальному) больше средней реакции самых молодых игроков. Ответ округлите до сотых.

```python
meanReaction_Older = football[football.Age == football.Age.max()].Reactions.mean()
meanReaction_Young = football[football.Age == football.Age.min()].Reactions.mean()
round(meanReaction_Older-meanReaction_Young, 3)
```

#### 8. Из какой страны (Nationality) происходит больше всего игроков, чья стоимость (Value) превышает среднее значение?

```python
football[football.Value > football.Value.mean()].Nationality.value_counts()
```

#### 9. Определите, во сколько раз средняя зарплата (Wage) голкипера (Position, GK) с максимальным значением показателя " Рефлексы" (GKReflexes) выше средней зарплаты голкипера с максимальным значением показателя "Владение мячом" (GKHandling). Ответ округлите до сотых.

```python
wage_1 = football[(football.Position == "GK") & (football.GKReflexes==football.GKReflexes.max())].Wage.mean()
wage_2 = football[football.GKHandling==football.GKHandling.max()].Wage.mean()
print(round(wage_1 / wage_2, 2))
```

#### 10. Определите, во сколько раз средняя сила удара (ShotPower) самых агрессивных игроков (игроков с максимальным значением показателя "Агрессивность" (Aggression)) выше силы удара игроков с минимальной агрессией.Ответ округлите до сотых.

```python
meanShotPowerMax = football[football.Aggression == football.Aggression.max()].ShotPower.mean()
meanShotPowerMin = football[football.Aggression == football.Aggression.min()].ShotPower.mean()
print(meanShotPowerMax / meanShotPowerMin)
```
