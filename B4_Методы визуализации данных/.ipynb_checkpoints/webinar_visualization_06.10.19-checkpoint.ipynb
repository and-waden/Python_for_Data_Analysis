```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings; warnings.simplefilter('ignore')
sns.set()
```

```python
df = pd.read_csv('train.csv')
```

```python
df.drop(['PassengerId', 'Ticket', 'Cabin'], axis = 1, inplace = True) # delete the columns
df['Fare'] = round(df['Fare'], 2) # round the price of tickets to 100
df['Embarked'].fillna('No data', inplace = True) # fill null values in Embarked column with value "No data"
```

```python
"""Fill null values of Pclass psngers"""

age_1 = df[df['Pclass'] == 1]['Age'].mean()
age_2 = df[df['Pclass'] == 2]['Age'].mean()
age_3 = df[df['Pclass'] == 3]['Age'].mean()

def fill_age(row):
    if pd.isnull(row['Age']):
        if row['Pclass'] == 1:
            return age_1
        elif row['Pclass'] == 2:
            return age_2
        else:
            return age_3
    return int(row['Age'])

df['Age'] = df.apply(fill_age, axis = 1)
```

```python
def is_alone(row):
    if row['SibSp']>0 or row['Parch']>0:
        return 'Не один'
    else:
        return 'Один'
    
df['alone'] = df.apply(is_alone, axis = 1)
```

### Рисуем несколько графиков в одном

```python
classes = df[df['Fare'] > 0].groupby(by = 'Pclass')['Fare'].agg(['min','max']).reset_index()
classes['Pclass'] = classes['Pclass'].apply(lambda x: {1:'Первый класс', 2:'Второй класс', 3:'Третий класс'}[x])
classes
```

```python
survived = df.groupby(by = 'Pclass')['Survived'].agg('mean').reset_index()
survived['Pclass'] = survived['Pclass'].apply(lambda x: {1:'Первый класс', 2:'Второй класс', 3:'Третий класс'}[x]) # x - имя ячейки к которой применяется lamda-функция
survived
```

```python
# Рисуем график, отображающий минимальную и максимальную стоимость билета по классам (в виде столбиковой диаграммы)
# и долю выживших по классам в виде линейного графика (все в одной системе координат)

fig = plt.figure()
axes = fig.add_axes([0, 0, 1.5, 1])
axes.set_ylim(0, 200) # apply limit to Y-axis via "ylim"
axes.bar(x = classes.Pclass, 
         height = classes["max"], 
         width = 0.4,
         align = "edge", # отцентровка столбиков
         label = "Макс стоимость билета")
axes.bar(x = classes.Pclass, 
         height = classes["min"], 
         width = -0.4,
         align = "edge",
         label = "Мин стоимость билета")
axes.legend()

axes_2 = axes.twinx()
axes_2.set_ylim(0, 1)
axes_2.plot(survived.Pclass, survived.Survived, color = "red")
axes_2.grid(axis = "y") # enhance background view
```
