## B2.4.1 Что такое сводные таблицы?


```python
import pandas as pd
```

```python
df = pd.read_csv("data_sf.csv")
df.head(10)
```

```python
pivot = df.loc[df['Club'].isin(['FC Barcelona','Real Madrid'
                                ,'Juventus','Manchester United'])].pivot_table(values=['Wage'], # 2.0 В параметр values мы передаем 
                                                                                                           # ту колонку, по которой нам нужно строить сводные данные, 
                                                                                                           # применяя агрегирующую функцию сумма.
# 1.0 df.loc[df['Club'].isin(['FC Barcelona','Real Madrid','Juventus','Manchester United'])] 
# 1.0 Она позволяет отфильтровать датафрейм по колонке Club, где значение в колонке соответствует одному 
# из перечисленных в массиве клубов.
                                                                                                                                                                                                                                                                                                           
index=['Nationality'], # В параметр index передадим ту колонку, данные которой будут представлены строками сводной таблицы.
columns=['Club'], # В параметр columns передаем колонку, значения которой будут в столбцах.
aggfunc='sum', # параметр aggfunc передаем агрегирующую функцию. 
margins=True) # Для того, чтобы добавить столбец и строку с итоговыми значениями, нужно добавить параметр margins=True:
display(pivot)
```

```python
df2 = df.pivot_table(columns = 'Position', index = 'Club', values = 'Wage', aggfunc = 'max')
df2
```

```python
df2.loc['Manchester City']['GK'] # максимальной зарплате вратаря (GK) в "Manchester City"?
```

```python
pivot.loc['Belgium'] 
```

```python
# Для того, чтобы заменить NaN на 0, можно применить дополнительный параметр fill_value.

pivot = df.loc[df['Club'].isin(['FC Barcelona','Real Madrid','Juventus','Manchester United'])].pivot_table(values=['Wage'],
index=['Nationality'],
columns=['Club'],
aggfunc='sum',
margins=True,
fill_value=0)
display(pivot)
```

#### B2.4.3.1 Задание 1
Создайте сводную таблицу, содержащую сведения о количестве игроков, 
занимающих разные позиции в каждом клубе. Отсутствующие значения замените нулями.

Каково среднее количество вратарей (GK) в клубе? Ответ округлите до трёх цифр после запятой

```python
pivot = df.pivot_table(values=['Name'], 
                       index=['Position'], 
                       columns=['Club'], 
                       aggfunc='count', 
                       fill_value=0)

mean_GK = round(pivot.loc['GK'].mean(),3) # Количество голкиперов в клубе
print(mean_GK)
```

#### B2.4.3.2 Задание 2

Используя таблицу, созданную на предыдущем шаге, определите, сколько клубов не содержат данных о центральных полузащитниках. (CM)

Подсказка: для выполнения этого задания желательно сохранить сводную таблицу в виде отдельного датафрейма и сгруппировать часть данных этого датафрейма с помощью value_counts().

```python
df3 = df.pivot_table(values = 'Name',
                           index = 'Club',
                           columns = 'Position',  
                           aggfunc = 'count',
                           #margins = True, #если указать - неправильно потом посчитает среднее
                                            # т.к. добавиться строка ALL с сумой всех игроков позиции
                           fill_value = 0)
                           
cm_cnt = df3['CM'].value_counts()
cm_cnt[0]
```

#### B2.4.4.1 Задача #1

С помощью сводной таблицы и функции loc посчитайте, сколько получают ("Wage") русские футболисты ("Russia"), играющие за ФК "AS Monaco".

```python
# 1 

pivot = df.pivot_table(values='Wage',
                       index='Nationality',
                       columns='Club', 
                       aggfunc='sum',
                       fill_value=0)

display(pivot.loc['Russia']['AS Monaco'])
```

```python
# 2

pivot = df.loc[df['Nationality'].isin(['Russia'])].pivot_table(values=['Wage'],
index=['Nationality'],
columns=['Club'],
aggfunc='sum',
# margins=True,
fill_value=0)
display(pivot.loc["Russia"][0])
display(pivot.loc["Russia"])
```

#### B2.4.4.2 Задание 2

Создайте сводную таблицу, содержащую сведения о средней скорости футболистов, занимающих разные позиции (Position) в разных футбольных клубах (Club).

Основываясь на данных таблицы, отметьте три позиции, представители которых в среднем обладают самой высокой скоростью.

```python
pivot = df.pivot_table(values='Acceleration',
                       index='Club',
                       columns='Position', 
                       aggfunc='max',
                       fill_value=0)

# display(pivot.head(5))
display(pivot.max().sort_values(ascending=False))
```

#### B2.4.4.3 Задание 3

Используя таблицу, созданную на предыдущем шаге, отметьте названия трёх клубов, в которых центральные форварды (ST) обладают наибольшей средней скоростью.

```python
pivot = df.pivot_table(values='Acceleration',
                       index='Position',
                       columns='Club', 
                       aggfunc='mean',
                       fill_value=0)

# display(pivot.loc["ST"].head(10))
display(pivot.loc["ST"].sort_values(ascending=False).head(10))
```
