#### B5.3.4 Очистка данных - Фильтрация 3

Очень часто приходится фильтровать значения в столбце по соответствию строке или ее части. 
Для этого есть две полезные функции

```python
# Обе функции не могут работать с NaN и необходимо использовать параметр na = False

str.match ("abc") – ищет строки которые начинаются c "abc"

str.contains ("abc") – ищет строки в которых есть "abc"
```

```python
sample.Name.str.match("К", na=False)
```

```python
sample[sample.Name.str.match("К", na=False)] 
# маска, наложенна на DF - все, что соответсвует значению в коде
# буква К
```

```python
sample[~sample.Name.str.match("К", na=False)] 
# маска, наложенна на DF - все, что НЕ соответсвует значению в коде
# буква К
```

```python
df.column_name.apply(func)

# Apply вызывает свою функцию для каждого элемента колонки. Функции, с которыми работает apply, бывают двух видов:

# лямбда-функции (можно задать внутри apply),
# обычные именные функции (определяются через def).
```

```python
# Возведём возраст в sample в квадрат:

sample.Age.apply(lambda x:x**2)
```

#### B5.3.6 Очистка данных - финальное задание
Соединим всё, что мы делали, и очистим файл log.csv. А именно:

прочитаем файл в переменную log;
добавим названия колонок user_id, time, bet, win;
удалим значения user_id с ошибками;
оставим в user_id только значения идентификатора;
уберём начальную скобку из поля time.

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

def user_num(userid):
    if userid == "#error":
        return ''
    
    else:
        return str(userid)[userid.find('-') + 2:]

log['user_id'] = log.user_id.apply(user_num)

def time_correct(x):
    return str(x)[1:] # убрать один символ из начала

log['time']=log['time'].str.replace('[', '')
log
```
