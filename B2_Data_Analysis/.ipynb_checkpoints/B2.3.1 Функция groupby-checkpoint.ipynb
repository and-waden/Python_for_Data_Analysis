## B2.3.1 Функция groupby

```python
import pandas as pd
```

```python
df = pd.read_csv("data_sf.csv")
df.head(5)
```

```python
df.groupby(['Club']) 
# Мы передали в функцию groupby на вход колонку (или список колонок), 
# по которой будет осуществляться группировка. 

# Это объект группировки, он хранит в себе информацию о том, 
# какие строки датафрейма (по индексным номерам) соответствуют определенной группе 
```

```python
# df.groupby(['Club']).groups

# Увидеть это можно, вызвав у объекта группировки атрибут groups
```

```python
grouped_df = df.groupby(['Club']).sum()
display(grouped_df)

# Получился датафрейм, где все параметры, к которым можно было 
# применить соответствующую агрегирующую функцию, были просуммированы
```

```python
grouped_df.loc['Ajax']
```

```python
grouped_df.loc['Ajax']['Wage']

# Теперь мы можем получить сумму зарплат игроков этого футбольного клуба:
```

```python
grouped_df = df.groupby(['Club'])['Wage'].sum()
display(grouped_df)

# ограничить список наших параметров и при группировке осуществлять 
# суммирование только по параметру "Wage" 
```

```python
grouped_df = df.groupby(['Club'])['Wage'].sum().sort_values(ascending=False)
print(grouped_df.head(5))

# отсортировать полученные данные по убыванию суммы зарплат с помощью функции sort_values:
```

#### B2.3.1.1 Задание 1
Отметьте позиции (Position), по которым общая сумма зарплат (Wage) игроков превышает 5 млн евро в год.

```python
grouped_df = df.groupby(['Position'])['Wage'].sum().sort_values(ascending=False)
print(grouped_df.head(10))
```

## B2.3.3 Другие агрегирующие функции

```python
df.groupby(['Nationality'])[['Club','Name']].nunique()

# Функция nunique, которая позволяет посчитать количество уникальных значений по серии. 
# Её лучше всего применять к тем колонкам датафрейма, в которых хранятся категорийные данные:
```

```python
df.groupby(['Club'])['Name'].count()

# Функция count, с помощью которой можно посчитать количество элементов в группе. 
# То есть результат её будет такой же, как у функции value_counts
```

```python
df.groupby(['Club'])['Dribbling'].median()
```

```python
df.groupby(['Club'])['Strength'].max()
```

```python
df.groupby(['Club'])['Balance'].min()
```

#### B2.3.3.1 Задание 1
Посчитайте среднюю (mean) и медианную (median) зарплату (Wage) футболистов из разных клубов (Club). В скольких клубах средняя и медианная зарплаты совпадают?

```python
df.groupby(['Club'])['Wage'].median()  # median
```

```python
df.groupby(['Club'])['Wage'].mean()  # mean
```

```python
# df.groupby(столбец_для_группировки)[столбцы_для_отображения].agg(['функция_1', 'функция_2'])

s = df.groupby(['Club'])['Wage'].agg(['mean', 'median'])
s=s.reset_index()
s.columns = ['Club','Mean','Median']
s.loc[s.Median == s.Mean].count()
# print(s)
```

## B2.3.3.2 Задание 2
Продолжаем работать с клубами, в которых средняя зарплата совпадает с медианной. Каков максимальный размер средней зарплаты в этой группе клубов?


```python
# df.groupby(['Club'])['Wage'].agg(['mean', 'median']).max().sort_values(ascending=False)
```

```python
s=df.groupby(['Club'])['Wage'].agg(['mean', 'median'])
s=s.reset_index()
s.columns = ['Club','Mean','Median']
s.loc[s.Median == s.Mean]#.count()
```
