#### Посчитайте объем выручки для каждой модели, которую можно получить, распродав все остатки на складе. Модель с каким item_id имеет максимальное значение выручки после распродажи остатков? Ответ дайте в виде целого числа.

```python
merged_dfs=items_df.merge(purchase_df, how='inner', on='item_id')
#display(s)
dct = {}

for i in range(len(merged_dfs)):
    d = merged_dfs['item_id'][i]
    f = merged_dfs['price'][i] * merged_dfs['stock_count'][i]
    dct[d] = [f]
    print("This is the ID: {}, this is the PRICE: {}".format(d, f))
    
# dct = pd.DataFrame(dct)
# display(dct)
# display(dct.sum().sort_values(ascending=False))
```
